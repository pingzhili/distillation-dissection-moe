```python

from matplotlib import pyplot as plt

plt.rcParams['font.family'] = 'Helvetica Neue'
```

```python
# of tokens with 0 different experts: 92588
# of tokens with 1 different experts: 71116
# of tokens with 2 different experts: 220543
# of tokens with 3 different experts: 218620
# of tokens with 4 different experts: 272875
# of tokens with 5 different experts: 248249
# of tokens with 6 different experts: 208728
# of tokens with 7 different experts: 138794
# of tokens with 8 different experts: 64311
different_routed_experts = [92588, 71116, 220543, 218620, 272875, 248249, 208728, 138794, 64311]
```

```python
# Draw the distribution of different routing experts with MIT colors
plt.figure(figsize=(4, 3))
plt.bar(range(9), different_routed_experts, color="#A31F34")

plt.xlabel("# of Routed Experts Difference")
plt.ylabel("# of Tokens")

# remove the grid
plt.grid(False)

# set y axis range
plt.ylim(0, max(different_routed_experts) + 10000)
plt.xlim(-0.5, 8.5)

ax = plt.gca()
for _, spine in ax.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(1)
    spine.set_edgecolor("#343a40")

plt.legend()
plt.show()
```
